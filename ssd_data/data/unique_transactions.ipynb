```python

to_decode_list = []
for i in range (50, 61):
    
    log_entries = []
    
    with open(f'ctf_exchange_neg_risk/logs_{i}M.csv', 'r') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            log_entries.append(row)

    for log in log_entries:
        topic = log[4]
        if (to_decode_list.__contains__(topic)):
            continue
        else:
            print(log)
            to_decode_list.append(topic)

print(f'Length of Unique Addresses: {len(to_decode_list)}')
print('\n'.join(map(str, to_decode_list)))
```
