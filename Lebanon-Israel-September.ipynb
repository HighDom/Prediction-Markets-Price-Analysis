The dataset was obtained directly from Polymarket, focusing on the prediction market titled: **"Israeli forces enters Lebanon in September"**. 

This market resolves based on official reports of Israeli ground forces entering Lebanon for combat between September 18 and 30, 2024.

# Comparison November vs September invasion probability

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('Lebanon-Israel-Price-Data.csv')
df2 = pd.read_csv('Israel-Lebanon Price-Data-November-Minutely.csv')

# Convert 'Timestamp (UTC)' to datetime
df1['Datetime'] = pd.to_datetime(df1['Timestamp (UTC)'], unit='s')
df2['Datetime'] = pd.to_datetime(df2['Timestamp (UTC)'], unit='s')

# Set 'Datetime' as the index
df1.set_index('Datetime', inplace=True)
df2.set_index('Datetime', inplace=True)

deadline_str = "2024-10-01T00:00:00Z"
deadline_datetime = pd.to_datetime(deadline_str)

# Filter data between '2006-09-30' and '2018-09-30'
start_date = '2024-09-30'
end_date = '2024-10-01'
df1_filtered = df1.loc[start_date:end_date]
df2_filtered = df2.loc[start_date:end_date]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df1_filtered.index, df1_filtered['Price'], label='Sep Invasion Price', color='red')
plt.plot(df2_filtered.index, df2_filtered['Price'], label='Nov Invasion Price', color='green')

plt.axvline(x=deadline_datetime, color='red', linestyle='-', linewidth=2)

plt.xlabel('Date (UTC)')
plt.ylabel('Price')
plt.title('Price Over Time (2024-09-30)')
plt.legend()
plt.show()
```
